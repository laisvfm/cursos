```python
#foi usado um prompt para traduzir as colunas para ðŸ‡µðŸ‡¹ BR
df = df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
})

df.columns
```

```python
df["senioridade"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
#foi usado um prompt para traduzir para ðŸ‡µðŸ‡¹ BR as categorias da coluna senioridade
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'FL': 'Freelancer',
    'PT': 'Meio PerÃ­odo'
})

df['contrato'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'MÃ©dia',
    'L': 'Grande',
    'S': 'Pequena'
})

df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'HÃ­brido',
    100: 'Remoto'
})

df['remoto'].value_counts()
```

```python
import numpy as np

#criaÃ§Ã£o de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'JoÃ£o', 'Maria', 'Carlos', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#calcula a mÃ©dia salarial e substitui os nulos pela mÃ©dia e arredonda os valores
df_salarios ['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#calcula a mediana e substitui os nulos pela mediana
df_salarios ['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda-feira', 'TerÃ§a-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()

df_temperaturas
```

```python
import pycountry
```

```python
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']

media_ds_pais = df_ds.groupby('residencia')['usd'].mean().sort_values(ascending=False).reset_index()

def codigo_para_nome(codigo):
    try:
        return pycountry.countries.get(alpha_2=codigo).name
    except:
        return codigo

media_ds_pais['nome_pais'] = media_ds_pais['residencia'].apply(codigo_para_nome)

fig = px.bar(media_ds_pais,
             x='usd',
             y='nome_pais',
             labels={'nome_pais': 'PaÃ­s', 'usd': 'SalÃ¡rio MÃ©dio (USD)'},
             title='SalÃ¡rio mÃ©dio de Cientista de Dados por paÃ­s',
             orientation='h',
             height=800)
fig.show()
```

```python
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com cÃ³digo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular mÃ©dia salarial por paÃ­s (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='SalÃ¡rio mÃ©dio de Cientista de Dados por paÃ­s',
                    labels={'usd': 'SalÃ¡rio mÃ©dio (USD)', 'residencia_iso3': 'PaÃ­s'})

fig.show()
```
